```python
arr = [1,0,2,3,0,4,5,0]


n = len(arr)

zeros = arr.count(0)

zeros

i = n -1

j = n-1 + zeros

i
j


while i < j:
    if j < n:
        arr[j] == arr[i]

    if arr[i]==0:
        j-=1
        if j < n:
            arr[j]= 0
    i -=1
    j-=1
```

```python
nums = [0, 1, 0, 3, 12]

write = 0

for read in range(len(nums)):
    if nums[read] != 0:
        nums[write] = nums[read]
        write += 1


while write < len(nums):
    nums[write] = 0
    write += 1

print(nums)
```
