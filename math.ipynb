```python
# recursion

# print some thing n times

def print_message(n):
    if n == 0:
        return
    
    print("hello world")

    print_message(n-1)


print_message(3)
```

```python
# print 1 to n

# 1->2->3->4->5

# but recursion goes from n->n-1->n-2->n-3

def print_1_to_n(n):
    if n == 0:
        return
    print_1_to_n(n-1)

    print(n)

print_1_to_n(5)
 
```

```python
# print n numbers in reverse

def print_n_to_1(n):
    if n == 0:
        return
    print(n)

    print_n_to_1(n-1)

print_n_to_1(5)

```

```python
def summation(n):
    if n == 0:
        return 0
    return n + summation(n-1)

summation(5)
```

```python
# reverse a array

# two pointers + recursion


def reverse(arr, left, right):
    if left >= right:
        return

    arr[left], arr[right] = arr[right], arr[left]

    reverse(arr, left + 1, right - 1)


arr = [1,2,3,4,5]

reverse(arr, 0, len(arr)-1)

arr
```

```python
arr = [1, 2, 2, 3, 1, 2, 4]

freq = {}

max_freq = 0

max_element = -1

for num in arr:
    freq[num] = freq.get(num, 0) + 1


for key, value in freq.items():
    if value > max_freq:
        max_freq = value
        max_element = key


print(freq)


print(max_element)

```
