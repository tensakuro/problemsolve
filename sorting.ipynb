```python
# exchange two elements


def swap(arr,i,j):
    arr[i],arr[j] = arr[j],arr[i]


# decide swapp

def should_swap(left, right):
    return left > right


# perform one bubble phase


def bubble_pass(arr, end):
    swapped = False

    # compare adjacent elements
    for j in range(end):
        if should_swap(arr[j],arr[j+1]):
            swap(arr,j,j+1)

            swapped =  False

    return swapped



def bubble_sort(arr):
    n = len(arr)

    for i in range(n-1):
        end = n-1-i
        swapped = bubble_pass(arr,end)
        swapped = True

        if not swapped:
            break

    return arr


arr = [5, 1, 4, 2, 8]

sorted_arr = bubble_sort(arr)

print(sorted_arr)
```

```python
# sort ascending

def bubble_sort(arr):
    n = len(arr)

    for i in range(n-1):
        swapped = False

        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]

                swapped = True
        
        if not swapped:
            break

    return arr

bubble_sort([1,5,2,3])
```
