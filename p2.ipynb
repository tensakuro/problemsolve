```python
def decorator(func):
    def wrapper(*args, **kwargs):
        print("before")
        result = func(*args, **kwargs)

        print("after")

        return result
    
    return wrapper


@decorator
def greet(name:str):
    print("hello", name)

greet("sohel")
```

```python
def require_login(func):
    def wrapper(user):
        if not user["logged_in"]:
            print("access denied")
            return

        return func(user)
    
    return wrapper

@require_login
def dashboard(user):
    print("welcome")


user1 = {"logged_in": True}

user2 = {"logged_in": False}


dashboard(user2)
dashboard(user1)

```

```python
def eligible_for_election(func):
    def wrapper(age: int):
        if age <18:
            print("not eligible")
            return
            
     

        return func(age)

    return wrapper


user1 = 19
user2 =17


@eligible_for_election
def vote(user1):
    print("voting enabled")


v = vote(user1)

print(v)


g = vote(user2)

print(g)

    
```
